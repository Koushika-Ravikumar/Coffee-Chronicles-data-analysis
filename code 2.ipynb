```python
# importing the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the Dataset
data = pd.read_csv('/content/drive/My Drive/coffee_analysis.csv')
data.head(10)
```

```python
# Data Cleaning
# finding the Missing Values

missing_values = data.isnull().sum()
print("Missing values in each column:\n",missing_values)
```

```python
# Removing the Missing Values

data_cleaned = data.dropna()
```

```python
# printing the number of null values in each column
print(data_cleaned.isnull().sum())
```

```python
# Removing the Duplicates

data_cleaned = data_cleaned.drop_duplicates()
```

```python
# Standardize column names (convert to lowercase)
data_cleaned.columns = data_cleaned.columns.str.lower()
```

```python
data_cleaned.head(10)
```

```python
# Filter data for United States only
data_cleaned = data_cleaned[data_cleaned['loc_country'] == 'United States']
```

```python
data_cleaned.head(10)
```

```python
# Removing Origin_1 column from dataframe
if 'origin_1' in data_cleaned.columns:
    # 3. Drop the 'origin_1' column
    data_cleaned = data_cleaned.drop(columns=['origin_1'])
    print("'origin_1' column has been removed.")
```

```python
data_cleaned.head(10)
```

```python
# Renaming Origin_2 to Origin
data_cleaned.rename(columns={'origin_2': 'origin'}, inplace=True)
```

```python
data_cleaned.head(10)
```

```python
# Get a summary of the data
summary = data_cleaned.describe()

# Display the summary statistics
summary
```

```python
# Calculate Pearson correlation
correlation = data_cleaned['100g_usd'].corr(data_cleaned['rating'])

# Display the correlation
print("Pearson correlation coefficient between Price and Ratings: " + format(correlation, ".3f"))
```

# Data Analysis <a class="anchor" id="step10"></a>

* [Zoom Up](#Contents)
* [Pseudocode](#step8)

### Q1: Most common coffee origins in the United States

```python
common_origins = data_cleaned['origin'].value_counts().head()
print("Most common coffee origins in the United States:")
print(common_origins)
```

### Bar chart for common coffee origins

```python
plt.figure(figsize=(10, 6))
common_origins.plot(kind='bar')
plt.title('Most Common Coffee Origins in US and Floyd')
plt.xlabel('Country of Origin')
plt.ylabel('Count')
plt.show()
```

#### Interpretation of the Bar-Chart
This bar chart shows the most common coffee origins in the US and Floyd.

- Colombia is the most common origin, followed by Southern Ethiopia and Guatemala.

- South-Central Kenya and Oromia Region also feature prominently, though with slightly lower counts compared to the top three origins.

### Q2: Average Ratings differ across various roast levels (light, medium, dark)

```python
ratings_by_roast = data_cleaned.groupby('roast')['rating'].mean()
print("\nAverage Ratings by Roast Level:")
print(ratings_by_roast)
```

###  Box-plot for ratings by roast level

```python
palette_colors = {"Medium-Light": "lightblue", "Medium": "Orange", "Light": "lightcoral","Medium-Dark":"Red","Dark":"Brown"}

plt.figure(figsize=(10, 6))
sns.boxplot(x='roast', y='rating', data=data_cleaned,palette=palette_colors)
plt.title('Ratings by Roast Level')
plt.show()
```

#### Interpratation of Box-Plot

This box plot shows the distribution of ratings for different coffee roast levels.

- Medium-Light and Light roasts tend to have higher median ratings compared to the other roast levels, with fewer outliers.
- Dark roast has the lowest median rating and a narrower spread, indicating lower variability in ratings.
- Medium and Medium-Dark roasts have a wider range of ratings, though their medians are still slightly below those of Light roasts.

### Q3: Most frequently mentioned origins in highly-rated reviews (threshold = 95)

```python
high_rating_threshold = 95
high_rated_reviews = data_cleaned[data_cleaned['rating'] >= high_rating_threshold]
most_mentioned_origins = high_rated_reviews['origin'].value_counts().head()
print("\nMost mentioned origins in highly-rated reviews:")
print(most_mentioned_origins)
```

### Word-cloud  for most mentioned origins

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(high_rated_reviews['origin'].dropna()))

# Plot the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Mentioned Origins in High-Rated Reviews')
plt.show()
```

#### Interpratation of the Word cloud
This word cloud shows the most frequently mentioned coffee origins in high-rated reviews.

- The larger and bolder the text, the more frequently that origin is mentioned in the reviews.
- Prominent origins include "Colombia," "Central," "Western," "Panama," "Kenya," and "Southern Ethiopia," indicating their strong presence in top-tier reviews.

### Q4: Price range for highly-rated coffees vs lower-rated coffees

```python
low_rating_threshold = 90  # Define lower rating threshold
high_rated = data_cleaned[data_cleaned['rating'] >= high_rating_threshold]
low_rated = data_cleaned[data_cleaned['rating'] < low_rating_threshold]

high_rated_price_stats = high_rated['100g_usd'].agg(['min', 'max', 'mean'])
low_rated_price_stats = low_rated['100g_usd'].agg(['min', 'max', 'mean'])

print("\nPrice range for highly-rated coffees:")
print(high_rated_price_stats)

print("\nPrice range for lower-rated coffees:")
print(low_rated_price_stats)
```

```python
low_rating_threshold = 90
low_rated_reviews = data_cleaned[data_cleaned['rating'] <= low_rating_threshold]
most_mentioned_origins = low_rated_reviews['origin'].value_counts().head()
print("\nMost mentioned origins in highly-rated reviews:")
print(most_mentioned_origins)
```

### Scatter-plot for price vs ratings

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='100g_usd', y='rating', hue='roast', data=data_cleaned)
plt.title('Price vs Ratings by Roast Level')
plt.show()
```

#### Interpratation of the Scatter-plot

This scatterplot shows the relationship between the price (100g in USD) and ratings of coffee by roast level.

- It indicates that higher-rated coffees tend to cluster between $0- $40 per 100g, with some exceptions at higher prices.
- The majority of highly rated coffees belong to the Medium-Light roast category, and the spread of ratings is fairly consistent across roast types, though Medium-Light and Medium dominate the range.
